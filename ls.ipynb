```python
import xml.etree.ElementTree as ET
import pandas as pd
import nltk
from nltk.tokenize import WordPunctTokenizer
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
from classes import lex_sub
```

```python
data = pd.read_csv("data/ag_news_preprocessed_train.csv").sample(frac=0.5, random_state=42)
```

```python
ls = lex_sub.LexSub()
```

```python
new_data = []
wpt = WordPunctTokenizer()

for i, row in data.iterrows():
  if i % 100 == 0:
    print(i)

  sentence = row["text"]
  label = row["label"]

  splitted = " ".join(wpt.tokenize(sentence))
  temp = []
  for w in splitted.split():
    r = ls.lexsub_concatenation(splitted, w)

    if r is None:
      temp.append(w)
    else:
      if len(r) > 0:
        temp.append(r[0][0])
      else:
        temp.append(w)
  new_data.append({"text":" ".join(temp), "label":label})
```

```python
new_df = pd.DataFrame(new_data)
new_df.to_csv("ls_concat_ag_news.csv", index=False)
```
